This notebook uses CPPTRAJ to save the frames from given trajectory. The following command is used to compute contacts in this notebook:
- trajout x_frame_reduced.pdb pdb onlyframes x 

```python
import os
```

#### The below cell contains all the inputs required from user. The inputs required are as follows:
1. Working diretory path
2. Topology file
3. Trajectory file
4. File containing list of frames separated by comma '__,__'
5. basename or output pattern for output (PDB files) 
6. Name for log file containing all the PDB file names generated by the script

```python
# All variable inputs are in this cell --> Change values accordingly
working_dir = '/home/hdhondge/MD_Musashi/Prev_Analysis/Annealing/5X3Z_simA/Analysis/'
top_file = 'topology_split.prmtop'  #Topology file --> path relative to working_dir
traj_file = 'reduced_cluster2.6.nc'  # Trajectory file with only frames you want to compute contacts for
frames_file = 'Abc'

basename_out = '_frame_reduced'
log_file = 'all_frames_saved.txt'
```

__The below cell will have all the necessary functions to create CPPTRAJ input files__.

```python
# funtion to create cpptraj input file
def create_in_cpptraj(topology, trajectory, frame, basename):
    with open(file, 'w') as f:
        f.write('parm ' + topology +'\n')
        f.write('trajin ' + trajectory + '\n')
        f.write('trajout ' + str(frame) + str(basename) + '.pdb pdb onlyframes ' +str(frame) + ' \n')
        f.write('run' + '\n')
    
    return file
```

__The actual code runs from below cell -->__

```python
os.chdir(working_dir)
os.system('conda activate MDsimu')
all_frames = open(frames_file).read().split(',')
logf = []

for a in all_frames:
    file = create_in_cpptraj(top_file, traj_file, a, basename_out)
    cmd = 'cpptraj -i '+ file 
    os.system(cmd)
    logf.append(str(a)+basename_out+ '.pdb')

# writing all output file names in a single file
with open(log_file, 'w') as f:
    f.write('\n'.join(logf))

print('Task completed successfully! Well done')
```
